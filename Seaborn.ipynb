# Seaborn

## Distribution plots
- distplot
- joinplot
- pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips  ')
```

```python
df.head() 
```

```python
#correlation with Heatmap  

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#distplot(univariate)

sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'],kde=True,bins=10)
```

```python
#joinplot

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#pair plot

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Categorical Plots
- boxplot
- violinplot
- countplot
- bar plot

```python
#boxplot

sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
#violinplot

sns.violinplot(x='total_bill',y='day',data=df)
```

```python
#countplot

sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y  ='sex',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
#bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```
