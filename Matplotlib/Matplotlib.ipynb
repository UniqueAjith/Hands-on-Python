```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#example
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#plt scatter
plt.scatter(x,y,c='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example for scatter')  
plt.savefig('Test.png') 
```

```python
#plt plot
y=x*x
```

```python
plt.plot(x,y, 'ro--')
```

```python
plt.plot(x,y, 'g*',linestyle='dashed',linewidth=2,markersize=12 )
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example for plot')
```

```python
#creating Subplots

plt.subplot(2,2,1)#(row,col,index)
plt.plot(x,y,'y--')
plt.subplot(2,2,2)
plt.plot(x,y,'go-') 
plt.subplot(2,2,3)
plt.plot(x,y,'r*')
plt.subplot(2,2,4)
plt.plot(x,y,'b.--')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,'g')
plt.show()
```

```python
#sin curve

x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title('Sin wave')
plt.plot(x,y,'r')
plt.show()
```

```python
# subplot of sin and cos wave

x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title('Sin Wave')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('Cos Wave')

plt.show()
```

```python
#Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')

plt.title('Bar plot Example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Histograms

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histograms title')
plt.show()
```

```python
#Box Plot

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data,vert=False,patch_artist=False);
```

```python
#pie chart

labels = 'HTML','CSS',"JS"
sizes = [215,120,245]
colors = ['gold','yellowgreen','lightskyblue']
explode = (0.1,0,0)
```

```python
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
